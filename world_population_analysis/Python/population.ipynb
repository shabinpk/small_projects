```python
#remove space from columns
df.columns=df.columns.str.strip().str.replace(" ","_")
```

```python
india_df=df.loc[df['Country/Territory']=='India']
#df.loc- label based indexed, it lets us select rows and columns using labeles,
#can be used to filter,slice or select based on conditions
```

```python
india_df
```

```python
population_2022=df.loc[:,['Country/Territory','2022_Population']]
population_2022
```

```python
population_2020_by_continent=df.groupby('Continent')['2020_Population'].sum()
population_2020_by_continent
```

```python
top10_populated_country_2022=df.sort_values(by='2022_Population',ascending=False).head(10)
top10_populated_country_2022
```

```python
avg_growth_by_continent=df.groupby('Continent')['Growth_Rate'].mean()
avg_growth_by_continent
```

```python
#reshape for time-series analysis by melting year columns into rows(long format)

df_long=df.melt(
    id_vars=['Country/Territory', 'Continent'],
    value_vars=['1970_Population','1980_Population','1990_Population',
                '2000_Population','2010_Population','2015_Population',
                '2020_Population','2022_Population'],
        var_name='Year',value_name='Population')
df_long
```

```python
world_trend=df_long.groupby('Year')['Population'].sum().reset_index()
world_trend
```

```python
df_long['Year']=df_long['Year'].str.replace("_Population","")
```

```python
df_long.dtypes
```

```python
df_long=df_long[['Year','Population']].astype(int)
```

```python
growth_rate_asia_vs_eu=df.groupby("Continent")["Growth_Rate"].mean().loc[["Asia","Europe"]].reset_index()
growth_rate_asia_vs_eu
```

```python
df
```

```python
world_population_perc_sumgrp_by_continent=df.groupby("Continent")['World_Population_Percentage'].sum().sort_values(ascending=False)
world_population_perc_sumgrp_by_continent.reset_index()
```

```python
df['pop_dif_1970vs2022']=df['2022_Population']-df['1970_Population']
df[['Country/Territory','pop_dif_1970vs2022']].sort_values('pop_dif_1970vs2022',ascending=False).head(10)
```
