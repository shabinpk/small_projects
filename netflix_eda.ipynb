```python
# since director take a good part of the data lets fill the NA of objects with Unknown instead of cleaning them(only date added Nan including rows will be removed)
df=df[df['date_added'].notna()] #returns a boolean mask (True/False for each row).df[...] filters rows where the condition is True. Means Keep only rows where date_added is not null
#df['director'].fillna('Not Available', inplace=True)
#df['cast'].fillna('Not Available', inplace=True)
#df['country'].fillna('Unknown', inplace=True)
#df['rating'].fillna('Unknown', inplace=True)
#df['duration'].fillna('Unknown', inplace=True)
#i am using loop on this
fill_values={
    'director': 'Not Available',
    'cast': 'Not Available',
    'country': 'Unknown',
    'rating': 'Unknown',
    'duration': 'Unknown'
}
for k,v in fill_values.items():
    df[k].fillna(v,inplace=True)
```

```python
#duration has min and seasons ..so we seperat them using string extract function (regex)
df[['duration_int','duration_type']]=df['duration'].str.extract(r'(\d+)\s*(\w+)')
#\d+ 1 or more digit extract (ex:90 from min,1,2 from seasons)
#s* 0 or more spaces
#\w+ 1or more word characters(letter/number/underscores) (ex:min,season,seasons)
#() groups the matches, (get two columns for 1:number 2:text)
```

### Visualization

```python
#Type Distribuion(Movies vs Shows)
type_counts=df['type'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_counts.index, type_counts.values,color=['orange','skyblue'])
plt.title("Count of Movies vs TV Shows")
plt.xlabel('Type')
plt.ylabel('Count')
plt.grid(axis='y',linestyle='--',alpha=0.6) 
#.grid used to add grid lines. #axis-y:add gridlines only to yaxis(horizontal)
#linestyle --:make the lines dashed
plt.show()
```

```python
#Average Movie Duration
movies=df[df['type']=='Movie'] #filters only movies for visualization
movies=movies.dropna(subset=['duration_int']) # drop rows with missing values..although i cleaned the data,just in case
avg_dur=movies['duration_int'].mean()
print("Average Movie Duration:",avg_dur)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(7,4))
sns.histplot(movies['duration_int'],bins=20,color='blue',edgecolor='brown')
plt.axvline(avg_dur,color='red',linestyle='--',label=f'Avg:{avg_dur:.1f} min') # Draw average line vertical on x axis
plt.title("Distribution of Movie Durations")
plt.xlabel("Duration (minutes)")
plt.ylabel("Number of Movies")
plt.legend()
plt.tight_layout() # automatically adjust spacing
plt.show()
```

```python
tv=df[df['type']=='TV Show'].copy()
tv['season_count']=tv['duration'].str.extract(r'(\d+)').astype(float) #extract season numbers from the duration column              
#\d  -any digit
#+ -one or more
#() - extract and capture in a group
#use copy() when creaing a slice
avg_seasons=tv['season_count'].mean()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(tv['season_count'],bins=15,color='mediumpurple')
plt.axvline(avg_seasons,color='red',linestyle='--',label=f'Avg:{avg_seasons:.1f} seasons')
plt.title('Distribution of TV Show Seasons')
plt.xlabel('Number of Seasons')
plt.ylabel('Number of Shows')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Barplot â€“ Average Movie Duration by Country (Top 5)

country_avg= movies.groupby('country')['duration_int'].mean().sort_values(ascending=False).head(5)
plt.figure(figsize=(10,5))
sns.barplot(y=country_avg.index,x=country_avg.values,color='green',alpha=0.5)
sns.set_theme(style='dark')
plt.title("Average Movie Duration by Top 5 Countries")
plt.ylabel("Countries")
plt.xlabel("Avg Duration")
plt.xticks(rotation=45) # to fix countries name overlapping
plt.show()
```

```python
#Boxplot â€“ Distribution of Movie Durations by Rating
plt.figure(figsize=(10,6))
sns.boxplot(data=movies,x='rating',y='duration_int',color='blue')
sns.set_theme(style='white')
plt.title("Distribution of Movie Durations by Rating")
plt.xlabel("Rating")
plt.ylabel("Duration")
plt.xticks(rotation=45)
plt.show()
```

```python
#Pie Chart â€“ Count of Movies by Rating(matplot)
rating_counts=movies['rating'].value_counts().head(5)
plt.figure(figsize=(8, 4))
plt.pie(rating_counts,labels=rating_counts.index,autopct='%1.1f%%',startangle=140)
plt.title("Top 5 Movie Rating Distribution")
plt.axis('equal')
plt.legend(
    loc='upper left', 
    bbox_to_anchor=(1, 1)  # moves legend outside
        )
plt.tight_layout()
plt.show()
```
